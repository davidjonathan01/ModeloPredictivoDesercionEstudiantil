# **Separación de las preguntas**

# **Analisis de las variables independientes con la dependiente**

```python
# Crear una figura y los subplots
fig, axes = plt.subplots(12, 2, figsize=(20, 50))

# Iterar sobre cada pregunta anterior
for i in range(12):
    # Obtener el nombre de la columna de la pregunta actual
    nombre_columna = df_preguntas.columns[i]

    # Calcular las frecuencias para la opción "Sí" en la pregunta actual
    frecuencias_si = df_preguntas[df_preguntas.iloc[:, 12] == 'Si'].iloc[:, i].value_counts()

    # Calcular las frecuencias para la opción "No" en la pregunta actual
    frecuencias_no = df_preguntas[df_preguntas.iloc[:, 12] == 'No'].iloc[:, i].value_counts()

    # Graficar para la opción "Sí" en la pregunta actual
    frecuencias_si.plot(kind='bar', color='#a9da88', ax=axes[i, 0])
    axes[i, 0].set_ylabel('Frecuencia')
    axes[i, 0].tick_params(axis='x', rotation=45)

    # Graficar para la opción "No" en la pregunta actual
    frecuencias_no.plot(kind='bar', color='lightcoral', ax=axes[i, 1])
    axes[i, 1].set_ylabel('Frecuencia')
    axes[i, 1].tick_params(axis='x', rotation=45)

    # Imprimir el nombre de la columna de la pregunta actual
    axes[i, 0].annotate(nombre_columna, xy=(0.25, 1), xytext=(0,5),
             xycoords='axes fraction', textcoords='offset points',
             size='large', ha='center', va='baseline')

    # Títulos de los gráficos
    axes[i, 0].set_title('Estudiantes que SÍ piensan desertar', pad=20, y=0.80)
    axes[i, 1].set_title('Estudiantes que NO piensan desertar', pad=20, y=0.80)

# Ajustar el espacio entre los subplots
plt.tight_layout()

# Mostrar los gráficos
plt.show()

```

# **Eliminación de preguntas no significativas**

```python
# Columnas a eliminar
columnas_a_eliminar = ['4. ¿Los horarios de los cursos son lo suficientemente flexibles en la FISI?',
                       '6. ¿Las responsabilidades familiares(como ser padre o madre , ser encargado de algún familiar , etc) afectan tu compromiso con tus estudios en la FISI?',
                       '9. ¿La falta de recursos universitarios que brinda la universidad , afectan o influyen tu rendimiento académico en tus estudios en la FISI?',
                       '11. ¿Cuántas horas a la semana dedicas a actividades extracurriculares(empleo, participación en grupos estudiantiles, etc)?'
                       ]
df_subset = df_preguntas.drop(columnas_a_eliminar, axis=1)
```

```python
df_subset
```

# **Conversión a variables dummies**

```python
#Feature engineering(one hot encoding)
df_subset_bin=pd.get_dummies(df_subset,drop_first=True)
df_subset_bin.sample(5)
```

```python
num_filas = df_subset_bin.shape[0]  # Número de filas
num_columnas = df_subset_bin.shape[1]  # Número de columnas

print("Número de filas:", num_filas)
print("Número de columnas:", num_columnas)
```

# **Creación y entrenamiento del modelo**

```python
#Test train split
x_train,x_test,y_train,y_test=train_test_split(df_subset_bin.drop('13. ¿Has considerado interrumpir tus estudios en la Facultad de Ingeniería de Sistemas e Informática (FISI) , como el cambio de carrera o la decisión de abandonar la universidad?_Si',axis=1),df_subset_bin['13. ¿Has considerado interrumpir tus estudios en la Facultad de Ingeniería de Sistemas e Informática (FISI) , como el cambio de carrera o la decisión de abandonar la universidad?_Si'])
```

```python
#train the model using the training data
LogReg=LogisticRegression(solver='liblinear')
LogReg.fit(x_train,y_train)
```

# **Rendimiento del modelo**

```python
#understanding the score
prediction=(LogReg.predict(x_test)>.5).astype(int)
np.sum(prediction == y_test) / len(y_test)
```

```python
df_subset_bin.sample(5)
```

# **Predicciones**

```python
# Realizar la predicción
prediccion = LogReg.predict(np.array([[0, 1, 1, 1, 1, 0, 1, 1, 1, 0]]))[0]

# Verificar si la predicción es igual a 1
if prediccion == 1:
    print("Es probable que el estudiante deserte")
else:
    print("No es probable que el estudiante deserte")
```

```python
#prediccion
prediccion=LogReg.predict(np.array([[0,1,1,1,1,1,0,0,0,1]]))[0]
# Verificar si la predicción es igual a 1
if prediccion == 1:
    print("Es probable que el estudiante deserte")
else:
    print("No es probable que el estudiante deserte")
```

# **Matriz de confusion**

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Hacer predicciones en el conjunto de prueba
y_pred = LogReg.predict(x_test)

# Calcular la matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)

# Mostrar la matriz de confusión como una imagen
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)

# Añadir etiquetas y títulos
plt.xlabel('Predicted labels')
plt.ylabel('Etiquetas verdaderas')
plt.title('Matriz de confusión')
plt.show()
```

# **Coeficientes de las variables**

```python
coeficientes_regresion = LogReg.coef_
coeficientes_regresion
```

```python
import matplotlib.pyplot as plt

coeficientes_regresion = LogReg.coef_[0]
nombres_variables = df_subset_bin.columns[:-1]  # Suponiendo que df es tu DataFrame de datos y la última columna es la variable de respuesta

plt.figure(figsize=(10, 6))
plt.barh(nombres_variables, coeficientes_regresion)
plt.xlabel('Coeficiente')
plt.ylabel('Variable')
plt.title('Coeficientes de Regresión Logística')
plt.grid(True)
plt.show()
```
